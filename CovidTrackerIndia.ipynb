```python
df=df.iloc[:,0:]
df.head(22)
```

```python
days=df[df['Current Status']=='Hospitalized'].groupby(['Month','Day'])[['Num Cases']].sum()
days
```

```python
len(days)
X=np.arange(len(days))
X
```

```python
y=days.values
y
```

```python
X=X.reshape(-1,1)
X
```

### Step 3: Dividing dataset into trainer and testing data for our regression model


```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures()
X_poly=poly.fit_transform(X)
poly.fit(X_poly,y)
reg_poly=LinearRegression()
reg_poly.fit(X_poly,y)
```

### Step 4: Plotting and data visualization


```python
plt.scatter(X,y,color='red')
plt.plot(X,reg.predict(X),color='blue')
plt.title('CovidTracker')
plt.xlabel('Days')
plt.ylabel('Number of Cases')
plt.show()
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,reg_poly.predict(poly.fit_transform(X)),color='blue')
plt.title('CovidTracker')
plt.xlabel('Days')
plt.ylabel('Number of cases')
plt.show()
```

```python
r_square=reg.score(X,y)
r_square
```

```python
rpoly_square=reg_poly.score(X_poly,y)
rpoly_square
```

```python
reg.predict([[250]])
```
